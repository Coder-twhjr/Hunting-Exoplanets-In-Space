####Check For The Missing Values

```
num_missing_values = 0

for column in exo_train_df.columns:
  for item in exo_train_df[column].isnull():
    if item == True:
      num_missing_values += 1

num_missing_values
```

```
non_missing_values = 0
for column in exo_train_df.columns:
  for item in exo_train_df[column].isnull():
    if item == False:
      non_missing_values += 1

non_missing_values
```
